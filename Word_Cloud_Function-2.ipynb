# Word Cloud Function 

```python
import numpy as np
import pandas as pd
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Function takes txt file stored in your Jupyter notebook and a list of words you don't want to include and makes 
# a wordcloud out of it
def make_wordcloud(txt_file, list_of_stopwords):
    df = pd.read_csv(txt_file, sep="\n", header=None)
    df.columns = ['tweets']
    text = " ".join(tweet for tweet in df.tweets)
    # Create stopword list (words to not include, in form ['this', 'that', 'the', 'of']
    stopwords = set(STOPWORDS)
    stopwords.update(list_of_stopwords)

    # Generate a word cloud image
    wordcloud = WordCloud(stopwords=stopwords, background_color="white").generate(text)

    # Display the generated image:
    # the matplotlib way:
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis("off")
    plt.show()
    
    # With Donald Trump's Face
    donald_mask = np.array(Image.open("donald2.png"))  # You will need to upload a silhouette picture of Trump's face
    # and save it in your Jupyter notebook as donald.png for this to work. Note that not all png's work.
    
    # Create a word cloud image
    wc = WordCloud(background_color="white", max_words=1000, mask=donald_mask,
               stopwords=stopwords, contour_width=3, contour_color='firebrick')

    # Generate a wordcloud
    wc.generate(text)
    
    
    # store to file
    wc.to_file("donald_bible_high_res.png") # Saves wordcloud to your files in Jupyter


make_wordcloud('First_Ten_Thousand.txt', ['the', 'of', 'unto', 'ye', 'let','upon', 'thanks', 'thank'])
```
